```python
def distanceMatrices(to, froom, mode="driving"):
    dm = gmaps.distance_matrix( to, froom,mode=mode, language=None, avoid=None, units="imperial",\
    departure_time=None, arrival_time=None, transit_mode=None,\
    transit_routing_preference=None, traffic_model=None)
    distanceMatrix = [[y['distance']['value'] for y in x['elements']] for x in dm['rows']]
    durationMatrix = [[y['duration']['value'] for y in x['elements']] for x in dm['rows']]
    return dm, distanceMatrix, durationMatrix
```

```python
import re
latLongs = []
location = []
import csv
with open('CPM_LAT_LONG.csv', 'rb') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        line = row[2] + "," + row[1]
#         cleanLine = re.sub('\W,.+','', line )
        clean2 = line.replace('\xc2\xa0','')
        latLongs.append(clean2)
        location.append(row[0])
```

```python
import time

rawMatrix = []
distMatrix = []
durMatrix = []
length = len(latLongs)
chunk = 10
sleepTime=0.05
n_full_chunks = int(length/chunk)
ii = 0
jj = 0
# n_full_chunks = 10
i = 0
while i <= n_full_chunks:
    j = 0
    if i == n_full_chunks:
        chunk1 = latLongs[i*chunk:length-1]
    else:
        chunk1 = latLongs[i*chunk:i*chunk+chunk]
    while j <= n_full_chunks:
        if j == n_full_chunks:
            chunk2 = latLongs[j*chunk:length-1]
        else:
            chunk2 = latLongs[j*chunk:j*chunk+chunk]
#         try:
        time.sleep(sleepTime)
        rawData, driveDist, driveDur = distanceMatrices(chunk1, chunk2, mode="walking")
        rawMatrix.append(rawData)
        distMatrix.append(driveDist)
        durMatrix.append(driveDur)
        print("Success at {} {}".format(i, j))
#         except:
#             print("Error at {} {}".format(i, j))
#             i == 9999999
#             j == 9999999
        j+=1
    i+=1
```
